```python
# condition 제거
train_a = train_a.drop(['A1-1','A1-2'],axis=1)
```

```python
# 1-3 : 0,1 카운트로 반응률을 변수로   #  반응          response
train_a['A1-3-0'] = train_a['A1-3'].apply(lambda x: x.count('0'))
train_a['A1-3-1'] = train_a['A1-3'].apply(lambda x: x.count('1'))

train_a['A1-3'] = train_a['A1-3-0']/(train_a['A1-3-0']+train_a['A1-3-1'])
train_a = train_a.drop(['A1-3-0','A1-3-1'], axis=1)
```

```python
# 1-4 : 평균값을 변수로 , 전체 데이터를 기준으로 75% 넘어가는 횟수 25% 넘어가는 횟수 그 사이 횟수를 변수로    # 반응시간          response
import numpy as np
import pandas as pd

# 문자열을 리스트(float)로 변환
train_a['A1-4'] = train_a['A1-4'].apply(lambda x: list(map(float, x.split(','))))

# 전체 반응시간을 모아 25%, 75% 분위값 계산
all_times = [t for sublist in train_a['A1-4'] for t in sublist]
q25 = np.percentile(all_times, 25)
q75 = np.percentile(all_times, 75)

print(f"25% 기준: {q25:.3f}, 75% 기준: {q75:.3f}")

# 각 행마다 fast/mid/slow count + 평균 계산 함수
def extract_a1_features(times):
    arr = np.array(times)
    fast = np.sum(arr < q25)
    slow = np.sum(arr > q75)
    mid = len(arr) - fast - slow
    mean = np.mean(arr)
    return pd.Series([fast, mid, slow, mean])

# 새로운 변수 추가
train_a[['A1_4_fast_cnt', 'A1_4_mid_cnt', 'A1_4_slow_cnt', 'A1_4_mean']] = (
    train_a['A1-4'].apply(extract_a1_features)
)

train_a = train_a.drop(['A1-4'], axis=1)
```

```python
# 타겟변수의 0과 1그룹에서 Cohen’s d가 낮고 다른변수들과 상관성이 낮으면(정보공유가 거의 없음) 변수 제거
#Cohen’s d는 두 그룹의 평균 차이를 표준편차 단위로 표현하는 효과크기(effect size)
import numpy as np

def cohens_d(group1, group2):
    # 평균
    mean1, mean2 = np.mean(group1), np.mean(group2)
    # 표준편차
    std1, std2 = np.std(group1, ddof=1), np.std(group2, ddof=1)
    # 풀드 표준편차 (pooled standard deviation)
    pooled_std = np.sqrt((std1 ** 2 + std2 ** 2) / 2)
    # Cohen's d
    d = abs(mean1 - mean2) / pooled_std
    return d

tr_a = train_a.drop(['Label'],axis=1)
column_list_a = []
for i in tr_a.columns:
  group0 = train_a[train_a['Label']==0][i]
  group1 = train_a[train_a['Label']==1][i]

  d = cohens_d(group0, group1)

  if d < 0.05:
      column_list_a.append(i)

```

```python
corr = train_a.corr()
remove_list_a = []

for var in column_list_a:
    if var == 'Label':
        continue

    # ① var의 앞 4글자가 포함된(같은 prefix 그룹의) 변수 찾기
    same_prefix_cols = [i for i in train_a.columns if var[:4] in i and i != var]

    # ② corr에서 해당 변수들과 자기 자신을 제외
    drop_cols = same_prefix_cols + [var]
    valid_cols = [c for c in corr.columns if c not in drop_cols]

    # ③ 나머지 변수들과의 상관계수 중 최대값 계산
    max_corr = corr[var].loc[valid_cols].abs().max()

    # ④ 기준 이하이면 제거 후보로 추가
    if max_corr < 0.5:
        remove_list_a.append(var)
```

```python
# 데이터셋 분할
import sklearn.model_selection as sklms
train_a_x = train_a.drop(['Label'],axis=1)
train_a_y = train_a['Label']

train_b_x = train_b.drop(['Label'],axis=1)
train_b_y = train_b['Label']


x_train_a,x_test_a,y_train_a,y_test_a = sklms.train_test_split(train_a_x,train_a_y,test_size=0.3,random_state=42,stratify=train_a_y)
x_train_b,x_test_b,y_train_b,y_test_b = sklms.train_test_split(train_b_x,train_b_y,test_size=0.3,random_state=42,stratify=train_b_y)
```

```python
# 모델 a
from xgboost import XGBClassifier
from sklearn.calibration import CalibratedClassifierCV

model_a = XGBClassifier(
    n_estimators=300,       # 트리 개수
    max_depth=50,           # 트리 깊이 (RF의 max_depth)
    learning_rate=0.005,     # 학습률
    scale_pos_weight=43,    # 불균형 보정 (Label 0:1 = 43:1)
    objective='binary:logistic',  # 이진 분류
    n_jobs=-1,
    random_state=42,
    tree_method='hist'      # 속도 빠른 GPU/CPU 버전
)

model_a.fit(x_train_a, y_train_a)
```

```python
calibrated_a = CalibratedClassifierCV(model_a, method='isotonic', cv=5)
calibrated_a.fit(x_train_a, y_train_a)
```

```python
import sklearn.metrics as sklm
import numpy as np

# 1) 확률 추출 (양성=1 클래스 확률)
proba_a = calibrated_a.predict_proba(x_test_a)[:, 1]

# 2) AUC / Brier → 확률로 계산
print("ROC-AUC :", sklm.roc_auc_score(y_test_a, proba_a))
print("Brier   :", sklm.brier_score_loss(y_test_a, proba_a))  # 낮을수록 좋음

# 3) 임계값 선택 (예: F1 최대값 기준)
prec, rec, th = sklm.precision_recall_curve(y_test_a, proba_a)
f1 = 2 * prec * rec / (prec + rec + 1e-12)
tau = th[np.argmax(f1[:-1])]  # 마지막 th 길이 보정
print("Best tau (by F1):", tau)

# 4) 분류 리포트는 레이블 필요 → 확률에 임계값 적용
pred_a = (proba_a >= tau).astype(int)
print(sklm.classification_report(y_test_a, pred_a, digits=3))
```

```python
# 데이터셋 분할
import sklearn.model_selection as sklms
train_a_x = train_a.drop(['Label'],axis=1)
train_a_y = train_a['Label']

train_b_x = train_b.drop(['Label'],axis=1)
train_b_y = train_b['Label']


x_train_a,x_test_a,y_train_a,y_test_a = sklms.train_test_split(train_a_x,train_a_y,test_size=0.3,random_state=42,stratify=train_a_y)
x_train_b,x_test_b,y_train_b,y_test_b = sklms.train_test_split(train_b_x,train_b_y,test_size=0.3,random_state=42,stratify=train_b_y)
```
